```python
class Usuario():
    '''
      Criar e armazenar dados dos usuarios de Python no Pythonbook.
    
         
    '''
    
    def __init__(self, nome, dat_nascimento, email):
        ''' Cronstrutor
                Criar objeto Pythonista a partir do nome, data de nascimento e e-mail.
                
            Parâmetros
                nome = str
            Nome do usuário.
                data_nascimento : str
            Data de nescimento do usuário no formato "dd/mm/aaaa".
                email = str
            Email do usuário 
        '''
        self.email = email
        self.nome = nome
        self.data_nascimento = dat_nascimento
        self.apelido = None
        self.lista_amigos = []
        self.lista_interesses = []
        self.posts = {}
        users_pythonbook.append(self)
   
    def __repr__(self):
        return f'{self.nome, self.email, self.data_nascimento}'
    
    def apelidar(self, apelido):
        '''
            Atribui ao objeto um valor ao campo apelido.
            
        Parâmtros
            apelido : str
            Apelido ao qual o usuário pode se denominar, funcionará para buscas na rede.
            
        '''
        self.apelido = apelido
        return True
    
          
    def verificar_existencia_usuario(self, usuario):# procurar por email 
        '''
            Busca o nome do objeto
            
          Parâmetros
          
              amigo : Pythonista()
              
                  Nome a ser buscado, digitado em minusculas e sem ''.
        '''
        if usuario in users_pythonbook:
            
            print('Usuário encontrado')
            return True
        else:
            print('Usuário não encontrado')
            return False
            

    def verificar_amigo_lista(self, amigo):
        '''
        Procura amigos lista salva pelo usuário
        
        Parâmetros
        amigo : Usuario()
            Deve ser escrito em minusculo
        '''
        if self.verificar_existencia_usuario(amigo) == True:        
            if amigo in self.lista_amigos:
                print(f'{amigo.nome} faz parte da lista amigos.')
                return True
            return False

    def adicionar_amigo(self, amigo):
        '''
        Salva amigo em lista de amigos.
        
        Parâmetros
        amigo : Usuario()
            Deve ser escrito em minusculo
        '''
        if self.verificar_amigo_lista(amigo) == False:
            self.lista_amigos.append(amigo)
            amigo.lista_amigos.append(self)
            print("Pronto, agora vocês são amigos")
            return True
        return False
            
    def adicionar_interesse(self, assunto):
        '''
            Salva os interessess em uma lista.
        
        Parâmetros
            assunto : str
            Deve ser escrito apenas um interesse por vez.
        '''
        if assunto in self.lista_interesses:
            print('Ok, esse assunto ja estava listado como seu interresse')
        else: 
            self.lista_interesses.append(assunto)
            print('Adicionado com sucesso')
    
    def postar(self, titulo, post):
        '''
        Salva em um dicionário os posts criados elo usuário.
        
        Parâmetro:
        titulo : str
        
            Titulo do post que será usado como chave do dicionário de posts
        post : str
        
            Postagem 
        '''
        self.posts[titulo] = post
    
    def numero_amigos(self):
        '''
        Imprime a quantidade de amigos salvos na lista de amigos.
        
        '''
        print(len(self.lista_amigos))
    
    def quantificar_posts(self):
        '''
        Imprime a quantidade de posts salvos.
        '''
        print(f'{len(self.posts)}')
    
    def deletar_post(self, titulo):
        '''
        Deleta o post usando o titulo para encontrar no dicionário.
        
        Parâmetro:
        
        titulo : str
        
        Título do post que será deletado.
        
        '''
        self.posts.pop(titulo)
        return 'Post deletado'
        
    def buscar_palavra(self, palavra):
        '''
        Busca uma palavra e retorna com o posta que a contém.
        
        Parâmetro
        palavra : str
            Palavra a ser procurada nos posts.
                
        '''
        print(f'A palavra "{palavra}" foi encontrada nos seguintes posts:')
        print('\n', self.nome, '\n')
        for titulo in self.posts:
                if palavra in titulo or self.posts[titulo]:
                    print (titulo,'\n' ,5*' '+self.posts[titulo])
        
        for nome in self.lista_amigos:
            
            print('\n',nome.nome, '\n')
            for titulo in nome.posts:
                if palavra in titulo or nome.posts[titulo]:
                    print (titulo,'\n' ,5*' '+nome.posts[titulo])
   
    def time_line(self):
        '''
        Imprime os posts de todos os amigos.
        '''
        for nome in self.lista_amigos:
            print('\n', nome.nome, '\n')
            for i in nome.posts:
                print(i, ':', 5*' ', nome.posts[i])
                
    def excluir_amigo(self, amigo):
        '''
        Exclui um amigo da lista de amigos.
        
        Parâmetro
        
        amigo : Usuario()
        
    
        '''
        
        if self.verificar_amigo_lista(amigo):
            self.lista_amigos.pop(self.lista_amigos.index(amigo))
            print('Usuário removido da lista de amigos.')
        
        
users_pythonbook = []
raphaela = Usuario('Raphaela', '19/06/1986', 'raphaelarps@gmail.com')
raphael = Usuario('Raphael', '19/06/1986', 'rapharps@gmail.com')
gabriel = Usuario('Gabriel', '21/03/1954', 'gabriel@gmail.com')
raphael.postar('primeiro', 'ola este é meu primeiro post.')
raphael.postar('segundo', 'como vai?')
raphael.postar('terceiro', 'IAI?')
gabriel.postar('Gato ','Miau')
gabriel.postar('Cachorro','Auau')
gabriel.postar('Papagaio','Da o pé loro')
raphaela.postar('x', 'y')
raphaela.postar('ab', 'cd')
raphaela.postar('DS','é vida')
gabriel.adicionar_amigo(raphael)
raphael.adicionar_amigo(raphaela)
```
